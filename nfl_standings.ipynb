```python

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Parse the game data

# Step 2: Create the graph
G = nx.DiGraph()

# Step 3: Add nodes and edges
for game in games:
    home_team = game["home_team"]
    away_team = game["away_team"]
    home_score = game["home_score"]
    away_score = game["away_score"]
    winner = home_team if home_score > away_score else away_team
    loser = away_team if winner == home_team else home_team
    margin = abs(home_score - away_score)
    
    # Add nodes
    if home_team not in G:
        G.add_node(home_team, points_diff=0)
    if away_team not in G:
        G.add_node(away_team, points_diff=0)
    
    # Update point differential
    G.nodes[home_team]['points_diff'] += (home_score - away_score)
    G.nodes[away_team]['points_diff'] += (away_score - home_score)
    
    # Add edge with the winner's color
    G.add_edge(loser, winner, weight=margin/5)

# Step 4: Normalize point differentials
points_diffs = [G.nodes[team]['points_diff'] for team in G.nodes]
min_diff = min(points_diffs)
max_diff = max(points_diffs)

# print(f"Point Differentials (min={min_diff}, max={max_diff}):")
# for team, diff in zip(G.nodes, points_diffs):
#     print(f"{team}: {diff}")

# Avoid division by zero if all point differentials are the same
if min_diff == max_diff:
    normalized_points_diffs = [0.5 for _ in points_diffs]
else:
    normalized_points_diffs = [(diff - min_diff) / (max_diff - min_diff) for diff in points_diffs]

# print("\nNormalized Point Differentials:")
# for team, norm_diff in zip(G.nodes, normalized_points_diffs):
#     print(f"{team}: {norm_diff}")

# Update the graph with normalized point differentials
for i, team in enumerate(G.nodes):
    G.nodes[team]['points_diff_norm'] = normalized_points_diffs[i]

# Step 4: Calculate initial rankings
team_ranks = {team: G.in_degree(team) for team in G.nodes}
# print("Initial Rankings:")
# for team, rank in sorted(team_ranks.items(), key=lambda x: x[1], reverse=True):
#     print(f"{team}: {rank}")

# Step 5: Adjust rankings recursively with weighted combination of wins and opponent strength
def recursive_rank(graph, team, team_ranks, depth=0, max_depth=4, win_weight=0.5, opponent_weight=0.5):
    if depth > max_depth:  # Limit recursion depth to prevent infinite loops
        return team_ranks[team]
    
    opponents = list(graph.successors(team)) + list(graph.predecessors(team))
    if not opponents:
        return team_ranks[team]
    
    opponent_strength = sum([recursive_rank(graph, opponent, team_ranks, depth + 1, max_depth, win_weight, opponent_weight) for opponent in opponents])
    average_opponent_rank = opponent_strength / len(opponents)
    
    # Weighted combination of wins and opponent strength
    return win_weight * team_ranks[team] + opponent_weight * average_opponent_rank

# Apply the recursive ranking adjustment
for team in team_ranks:
    team_ranks[team] = recursive_rank(G, team, team_ranks)


# print("\nFinal Rankings:")
# for team, rank in sorted(team_ranks.items(), key=lambda x: x[1]):
#     print(f"{team}: {rank}")

# Normalize ranks for color mapping
min_rank = min(team_ranks.values())
max_rank = max(team_ranks.values())
norm_ranks = {team: (rank - min_rank) / (max_rank - min_rank) for team, rank in team_ranks.items()}

# Step 6: Visualize the graph
initial_pos = {team: (np.random.rand(), norm_ranks[team]) for team in G.nodes}

pos = nx.spring_layout(G, pos=initial_pos, seed=42, k=8, iterations=10)
# for team in pos:
#     pos[team][1] = norm_ranks[team]
node_colors = [plt.cm.coolwarm(norm_ranks[team]) for team in G.nodes]
node_sizes = [((G.nodes[team]['points_diff_norm']+0.5) * 15)**3 for team in G.nodes]
team_colors = {team: color for team, color in zip(G.nodes, node_colors)}
edge_colors = [team_colors[edge[1]] for edge in G.edges]
edge_widths = [G[u][v]['weight'] for u, v in G.edges]

plt.figure(figsize=(12, 12))
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=node_sizes, edge_color=edge_colors, width=edge_widths, arrows=True, connectionstyle='arc3,rad=0.2',)
plt.savefig("nfl_graph.png")
```
